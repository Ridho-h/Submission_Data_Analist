## Exploratory Data Analysis (EDA)

### Explore orders_df, order_items_df, products_df, dan product_category_names_df

```python
orders_df.describe()
```

```python
orders_df.sort_values(by="order_purchase_timestamp", ascending=False)
```

**Membuat df yang berisi orders dalam 3 bulan terakhir(Bulan 7, 8, 9 pada tahun 2018)**

```python
orders_df["order_purchase_timestamp"] = pd.to_datetime(orders_df["order_purchase_timestamp"])
new_orders_df = orders_df[(orders_df["order_purchase_timestamp"].dt.year == 2018) & (orders_df["order_purchase_timestamp"].dt.month.isin([7, 8, 9]))]
new_orders_df.head()
```

```python
order_name_items_df = pd.merge(
    left=orders_df,
    right=order_items_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
order_name_items_df.head()
```

```python
new_order_name_items_df = pd.merge(
    left=new_orders_df,
    right=order_items_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
```

```python
new_order_name_items_df.head()
```

```python
product_name_translated_df = pd.merge(
    left=products_df,
    right=product_category_names_df,
    how="left",
    left_on="product_category_name",
    right_on="product_category_name"
)
product_name_translated_df.head()
```

```python
new_order_item_name_translated_df = pd.merge(
    left=new_order_name_items_df,
    right=product_name_translated_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
new_order_item_name_translated_df.head()
```

```python
order_item_name_translated_df = pd.merge(
    left=order_name_items_df,
    right=product_name_translated_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
order_item_name_translated_df.head()
```

**Jumlah order berdasarkan kategori barang**

```python
sum_product_df = new_order_item_name_translated_df.groupby(by="product_category_name_english").order_item_id.sum().sort_values(ascending=False).reset_index()
sum_product_df
```

### Explore order_reviews_df

```python
order_reviews_df.head()
```

```python
all_df = pd.merge(
    left=order_item_name_translated_df,
    right=order_reviews_df,
    how="inner",
    left_on="order_id",
    right_on="order_id"
)
all_df.head()
```

```python
positive_reviews_df = order_reviews_df[(order_reviews_df["review_score"] == 4) | (order_reviews_df["review_score"] == 5)]
positive_reviews_df.head()
```

```python
new_order_item_positive_reviews_df = pd.merge(
    left=new_order_item_name_translated_df,
    right=positive_reviews_df,
    how="inner",
    left_on="order_id",
    right_on="order_id"
)
order_item_positive_reviews_df.head()
```

**Insight:**
- Kategori "health_beauty" dan "bed_bath_table" menjadi kategori dengan penjualan terbanyak dalam tiga bulan terakhir.
- Terdapat 11547 penjualan dengan review positive dalam tiga bulan terakhir.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Produk apa yang memiliki penjualan tertinggi dan terendah dalam tiga bulan terakhir?

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))
 
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
 
sns.barplot(x="order_item_id", y="product_category_name_english", data=sum_product_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Most Purchased Product", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)
 
sns.barplot(x="order_item_id", y="product_category_name_english", data=sum_product_df.sort_values(by="order_item_id", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Least Purchased Product", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)
 
plt.suptitle("Most and Least purchased Product by Number of Sales", fontsize=20)
plt.show()
```

### Pertanyaan 2: Bagaimana tren penjualan bulanan pada produk perawatan kulit ?

```python
beauty_categories = ['perfumery', 'health_beauty', 'cosmetics']
beauty_overall_data = all_df[all_df['product_category_name_english'].isin(beauty_categories)]
beauty_overall_sales_trend = beauty_overall_data.resample('ME', on='order_purchase_timestamp').size()


plt.figure(figsize=(10, 6))
plt.plot(beauty_overall_sales_trend.index, beauty_overall_sales_trend.values, marker='o', color='teal')
plt.title('Overall Sales Trend of Beauty Products', fontsize=15)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Number of Sales', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Insight:**
- Kategori produk yang paling banyak dibeli adalah "health_beauty" sedangkan yang paling sedikit adalah "furniture_mattress_and_uphoistery".
- Penjualan Produk Kecantikan mengalami peningkatan drastis pada bulan Desember 2017 dan tidak bertahan lama mengalami penurunan drastis pada bulan berikutnya atau Januari 2018.
